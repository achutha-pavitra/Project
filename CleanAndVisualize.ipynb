```python
### import necessary packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
### read csv using pandas
house = pd.read_csv('./train.csv')
```

```python
### read first few lines to check the data that tells we have 81 columns and 5 rows.
house.head()
```

```python
### get more information using info method. This gives basic picture of the data type of the high level info of columns and counts
house.info()
```

```python
### check if any null values ### there are around 19 columns with null values
pd.options.display.max_rows=None
house.isnull().sum()
```

```python
### replacing null values with mean number ###
house.loc[house['LotFrontage'].isnull()==True, 'LotFrontage'] = 70.049
```

```python
### Most of them are not NAs. NAs are added to represent No Alley, No fireplace, No Pool, No Fence, No miscfeature. 
### It doesn't mean the data is unavailable
### We will replace these numbers with merging other datasets with corresponding values and delete the original columns after
house.isnull().sum()
```

```python
### read csv's using pandas These datasets are manually made to match the original datasets to replace NAs
alley = pd.read_csv('./Alley.csv')
fence = pd.read_csv('./Fence.csv')
fireplace = pd.read_csv('./Fireplace.csv')
misc = pd.read_csv('./Misc.csv')
pool = pd.read_csv('./Pool.csv')
garage=pd.read_csv('./garage.csv')
garagecond=pd.read_csv('./garagecond.csv')
garagefinish=pd.read_csv('./garagefinish.csv')
garagequal=pd.read_csv('./garagequal.csv')
```

```python
### Merge datasets with NAs with respective descriptive ones###
house_merge = pd.merge(house, alley, on='Alley', how='outer')
```

```python
house_merge = pd.merge(house_merge, fence, on='Fence', how='outer')
```

```python
house_merge = pd.merge(house_merge, fireplace, on='FireplaceQu', how='outer')
```

```python
house_merge = pd.merge(house_merge, misc, on='MiscFeature', how='outer')
```

```python
house_merge = pd.merge(house_merge, pool, on='PoolQC', how='outer')
```

```python
house_merge = pd.merge(house_merge, garage, on='GarageType', how='outer')
```

```python
house_merge = pd.merge(house_merge, garagecond, on='GarageCond', how='outer')
```

```python
house_merge = pd.merge(house_merge, garagefinish, on='GarageFinish', how='outer')
```

```python
house_merge = pd.merge(house_merge, garagequal, on='GarageCond', how='outer')
```

```python
### we can drop the other duplicate columns Alley, fireplace, Pool, Fence, miscfeature
house_merge.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond'], axis=1, inplace=True)
```

```python
### check once again to see if NAs
house_merge.isna().sum()
```

```python
### After analysis we are good to drop rest of the NA's
house_merge = house_merge.dropna()
```

```python
### All good to proceed with further analysis
house_merge.isna().sum()
```

```python
pd.options.display.max_columns=None
house_merge.head()
```

```python
### create a count plot to see the HouseStyle counts
sns.set_style("darkgrid")
g = sns.countplot(x="HouseStyle", data=house_merge)
g.set(title="Number of House style" )
#
plt.show()
```

```python
### create a scatterplot to see the Sale price Vs HouseStyle
g = sns.scatterplot(x="SalePrice", y="HouseStyle", data=house_merge, hue="CentralAir", style="CentralAir", hue_order=["Y", "N"])
g.set(title="Sale Price Vs House Style", xlabel="Sale Price", ylabel="House Style")
plt.show()
```

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Grvl": "green", "Pave": "blue"}
```

```python
### count plot to see the street type and most of them are Paved streets
g = sns.countplot(x="HouseStyle", data=house_merge, hue="Street", palette=palette_colors)
g.set(title="Number of House Style on diff Street type", xlabel="House Style", ylabel="Count")
plt.show()
```

```python
### Using relplots to visualize Change to use relplot() instead of scatterplot()
g = sns.relplot(x="OverallCond", y="SalePrice", 
                data=house_merge, kind="scatter", col="LotShape")
g.set(title="Overall Condition vs Sale prices on diff shape of property", xlabel="Over all condition", ylabel="Sale Price")
# Show plot
plt.show()
```

```python
# Create line plot
g = sns.relplot(x="YrSold", y="SalePrice", data=house_merge, kind="line", ci=None, style="HouseStyle", hue="HouseStyle",
            markers=True, dashes=False)
plt.xticks(rotation=45)
g.set(title="Year Sold Vs Sale Price", xlabel="Year of sold", ylabel="Sale Price")

# Show plot
plt.show()
```

```python
# Create count plot 
g = sns.catplot(x="Foundation", data=house_merge, kind="count")
g.set(title="Number of Foundation types", xlabel="Type of foundation", ylabel="Overall count")
plt.show()
```

```python
# Create a bar plot 
sns.set_style("whitegrid")
g = sns.catplot(x="HouseStyle", y="OverallQual", data=house_merge, kind="bar", ci=None)
g.set(title="House Style vs Over All Quality", xlabel="House style", ylabel="Over all quality")

# Show plot
plt.show()
```

```python
# Create a box plot
# Set the style to "darkgrid"
sns.set_style("darkgrid")
g = sns.catplot(x="RoofStyle", y="OverallCond", data=house_merge, kind="box", sym="")
g.set(title="Different Roof Style Vs Over all Condition", xlabel="Roof style", ylabel="Over all condition")
plt.show()
```
